```python
import os
import numpy as np
from astropy.io import fits
from astropy import units as u
from astropy.wcs import WCS
import scopesim as sim
from matplotlib import pyplot as plt
from matplotlib.colors import LogNorm
%matplotlib inline
import datetime
import shutil
import scopesim_templates as sim_tp
```

### Galaxie with METIS and MICADO


```python
gal = sim_tp.extragalactic.galaxies.spiral_two_component(extent=16*u.arcsec, fluxes=(15, 15)*u.mag)
```

```python
#HST telesxcope
cmd_H = sim.UserCommands(use_instrument="HST")

hst = sim.OpticalTrain(cmd_H)
hst.observe(gal,properties={"!OBS.dit": 100},
             update=True)
hdus_hst = hst.readout()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
plt.imshow(hdus_hst[0][1].data[345:385, 525:565], norm=LogNorm(), origin="lower")
plt.colorbar()
plt.title("HST")
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
plt.imshow(hdus_hst[0][1].data, norm=LogNorm(), origin="lower")
plt.colorbar()
plt.title("HST")
```

```python
wave = np.linspace(0.3, 2.5, 2201) * u.um
plt.figure(figsize=(15, 10))

plt.subplot(221)
plt.imshow(gal.fields[0].data, origin='lower')
plt.subplot(222)
plt.plot(wave, gal.spectra[0](wave))
plt.subplot(223)
plt.imshow(gal.fields[1].data, origin='lower')
plt.subplot(224)
plt.plot(wave, gal.spectra[1](wave))
```

```python
cmd = sim.UserCommands(use_instrument="MICADO", set_modes=["MCAO", "IMG_4mas"])
```

```python
cmd = sim.UserCommands(use_instrument="MICADO", set_modes=["MCAO", "IMG_4mas"])

cmd.update(properties={
    "!OBS.filter_name_fw1": "open",
    "!OBS.filter_name_fw2": "H",
    "!OBS.ndit": 20,
    "!OBS.dit": 3,

    "!OBS.catg": "SCIENCE",
    "!OBS.tech": "IMAGE",
    "!OBS.type": "OBJECT",
    "!OBS.mjdobs": datetime.datetime(2022, 1, 1, 2, 30, 0)
})

cmd["!DET.width"] = 4096     # pixel
cmd["!DET.height"] = 4096

cmd.update(properties = {"!Obs.dit" : 100})
cmd["!DET.width"] = 4096     # pixel
cmd["!DET.height"] = 4096

micado = sim.OpticalTrain(cmd)

micado.observe(gal)
hdus = micado.readout()

plt.figure(figsize=(10,10))
im = hdus[0][1].data 
#vmax=1.1*np.median(im), vmin=0.99*np.median(im)

plt.imshow(hdus[0][1].data, norm=LogNorm(), origin='lower')

#vmin=1e4
plt.colorbar()
```
