```python
import numpy as np
import os
import time
import matplotlib.pyplot as plt
import scipy
import matplotlib
import sympy
import numba
from numba import njit, prange , set_num_threads, get_num_threads
from typing import Callable
```

### Exercise 1

```python
def get_acceleration(X: np.ndarray) -> np.ndarray:
    N = X.shape[0]
    a = np.zeros_like(X)  # Initialize acceleration array
    
    for i in range(N):
        for j in range(N):
            if i != j:
                r_ij = X[j] - X[i]  # Vector from i to j
                distance = np.linalg.norm(r_ij)
                if distance > 0:
                    a[i] += r_ij / distance**3  # Gravitational acceleration formula
    
    return a
```

We create an array $a$ in the shape of (N, 3) that is filled with zeros to store the accelerations. We then iterate over each body $i$ to compute its acceleration. Inside the loop, we iterate over all other bodies $j$ (exlcuding $i$) where we compute the displacement vector $r_{ij} = X[j] - X[i]$, which points from body $i$ to body $j$. We then distance $|r_{ij}|$ between the two bodies. We the cube the inverse of the distance to match the denominator of the gravitational formula. We sum up all the contributions from the other bodies and store the final acceleration in $a[i]$. The result is a $(N, 3)$ array containing the acceleration vectors for all bodies.

### Exercise 2

```python
def measure_execution_time(N_values):
    times = []
    for N in N_values:
        X = np.random.randn(N, 3)  # Generate random positions
        start_time = time.perf_counter()
        get_acceleration(X)
        end_time = time.perf_counter()
        times.append(end_time - start_time)
    return times
```

```python
def plot_scaling(N_values, times):
    plt.figure()
    plt.loglog(N_values, times, marker='o', linestyle='-', label='Measured time')
    plt.loglog(N_values, (np.array(N_values)**2) * (times[0] / N_values[0]**2), linestyle='--', label='O(NÂ²) reference')
    plt.xlabel('Number of bodies (N)')
    plt.ylabel('Execution Time (s)')
    plt.title('Scaling of Gravitational Acceleration Computation')
    plt.legend()
    
    # Fit a line to the log-log data to get the slope
    coeffs = np.polyfit(np.log(N_values), np.log(times), 1)
    slope = coeffs[0]
    print(f"Estimated slope: {slope:.2f}")
    
    plt.show()
```

```python
# Run and plot results
N_values = [10, 20, 50, 100, 200, 500, 1000]  # Different values of N to test
execution_times = measure_execution_time(N_values)
plot_scaling(N_values, execution_times)
```

To verify the $O(N^2)$ complexity, we generated different values of $N$ that were ranging from $10$ to $1000$. For each value we generated random positions. We then measured the execution time of the function for the previous exercise for different values of $N$. We then plotted a log-log graph of computation time vs. $N$ to visualize the complexity. Additionally, we included a reference line representing the $O(N^2)$ scaling. We fitted a line to the log-log data and extracted the slope, which represents the complexity of the algorithm. The slope is $1.88$ which very close to $2$ which is as expected and confirms the behaviour of $O(N^2)$. The slope represents how computation time scales with $N$. A slope of $2$ means that doubling $N$  results in roughly $4$ times the computation time, which matches the expected $O(N^2)$. Our measured slope of $1.88$ is very close to $2$, indicating that while there may be small deviations due to implementation details or hardware effects, the algorithm indeed scales quadratically as expected.




### Exercise 3

```python
@njit
def get_acceleration_numba(X: np.ndarray) -> np.ndarray:
    N = X.shape[0]
    a = np.zeros((N, 3))
    
    for i in range(N):
        for j in range(N):
            if i != j:
                r_ij = X[j] - X[i]
                dist = np.linalg.norm(r_ij)
                if dist > 0:
                    a[i] += r_ij / dist**3
    
    return a
```

```python
def measure_execution_time(N_values, func):
    times = []
    for N in N_values:
        X = np.random.randn(N, 3)  # Generate random positions
        start_time = time.perf_counter()
        func(X)
        end_time = time.perf_counter()
        times.append(end_time - start_time)
    return times
```

```python
def plot_scaling(N_values, times_numpy, times_numba):
    plt.figure()
    plt.loglog(N_values, times_numpy, marker='o', linestyle='-', label='NumPy time')
    plt.loglog(N_values, times_numba, marker='s', linestyle='-', label='Numba time')
    #plt.loglog(N_values, (np.array(N_values)**2) * (times_numpy[0] / N_values[0]**2), linestyle='--', label='O(NÂ²) reference')
    plt.xlabel('Number of bodies (N)')
    plt.ylabel('Execution Time (s)')
    plt.legend()
    plt.title('Scaling of Gravitational Acceleration Computation')
    plt.show()
```

```python
def compare_methods():
    N_values = [10, 20, 50, 100, 200, 500, 1000]  # Different values of N to test
    times_numpy = measure_execution_time(N_values, get_acceleration)
    times_numba = measure_execution_time(N_values, get_acceleration_numba)
    
    speedup_factors = np.array(times_numpy) / np.array(times_numba)
    print("Speedup factors (NumPy / Numba):", speedup_factors)
    
    plot_scaling(N_values, times_numpy, times_numba)

compare_methods()
```

Here we took the function from exercise $1$ and added @njit to it from Numba, which allows it to run with Just-In-Time (JIT) compilation for better performance. To compare runtime performance, we reused the timing measurement function from Exercise 2, but modified it to accept different acceleration functions. This allowed us to test both the NumPy and Numba versions side by side. 

We also updated the scaling plot to include both methods, removing the O(NÂ²) reference line since we were focusing on relative performance rather than theoretical complexity. Additionally, we created a function to compute the speedup factor, which is the ratio of execution times between NumPy and Numba for each N. The results were then plotted for visualization.
The measured speedup factors ranged from 9 to 22 times faster, showing that Numba significantly outperforms NumPy. The plot confirms this, illustrating that the Numba method is consistently much faster across all tested values of N.

Numba is faster than NumPy because Numba compiles the function to efficient machine code, avoiding Pythonâs runtime overhead. Pure NumPy relies on vectorized operations, which work well but can struggle with nested loops like our $O(N^2) algorithm. With @njit, Numba optimizes explicit for-loops, making them almost as efficient as C or Fortran implementations. Numba reduces unnecessary memory allocations, while NumPy may create intermediate arrays that slow things down.

### Exercise 4

```python
def manual_validation(a1: np.ndarray, a2: np.ndarray, atol: float = 1e-6) -> bool:
    if a1.shape != a2.shape:
        return False  # Arrays must have the same shape

    for i in range(a1.shape[0]):  # Loop over bodies
        for j in range(a1.shape[1]):  # Loop over x, y, z components
            if abs(a1[i, j] - a2[i, j]) > atol:
                print(f"Mismatch found at ({i}, {j}): {a1[i, j]} vs {a2[i, j]}")
                return False  # Found a mismatch

    return True  # All values are close within tolerance
```

```python
def validate_acceleration(method1: Callable, method2: Callable, X: np.ndarray) -> bool:
    a1 = method1(X)  
    a2 = method2(X)  

    # Use our manual validation method
    manual_result = manual_validation(a1, a2)
    
    # Compare to numpy.allclose
    numpy_result = np.allclose(a1, a2, atol=1e-6)

    print(f"Manual validation result: {manual_result}")
    print(f"numpy.allclose result: {numpy_result}")

    return manual_result
```

```python
N_values = [10, 20, 50, 100, 200, 500, 1000]

for N in N_values:
    X = np.random.rand(N, 3)  # Random positions for N bodies
    print(f"\nValidating for N={N}:")
    validate_acceleration(get_acceleration, get_acceleration_numba, X)
```

To validate the resulting accelartions from the numba method, we need to compare them to the resulting accelerations from the numpy method. For that we again compute the accelerations using both methods. We then use our "manual_validation" function to check if the two acceleration arrays are nearly identical, with an absolute tolerance of $1e-6$. We use absolute tolerance because floating-point calculations can introduce small rounding errors. 

In our function we first check if both arrays have the same shape. If not, they are definitely different. We loop over a1.shape[0] (the number of bodies) to compare acceleration vectors body by body. We then loop over a1.shape[1] (which is always 3, representing the x, y, and z components) to compare each component individually. $a1$ and $a2$ are both (N, 3) arrays, where N is the number of bodies in our simulation. Using a1.shape[0], we ensure we iterate over each body ($i$ index). This allows us to compare the acceleration vectors body by body. Each body's acceleration has 3 components: $ax$, $ay$, and $az$. Using a1.shape[1], we ensure we iterate over each component ($j$ index). This allows us to compare the acceleration vectors component by component.

We compare each corresponding element in a1 and a2. If the absolute difference is larger than the absolute tolerance, we print the mismatch and return False. If the absolute difference is smaller than or equal to atol, it means the two results are numerically close enough, and we consider them equal within the allowed tolerance.

To compare the result from our funtcion with allclose, we create another function "validate_accelerarion". This function takes two functions and an input array. It computes the acceleration values using both methods and compares them. It uses our manual check and also allclose for comparison. It then prints both results to see if they match. In our case all of them match.

### Exercise 5

```python
@njit(parallel=True)
def get_acceleration_parallel(X: np.ndarray) -> np.ndarray:
    N = X.shape[0]
    a = np.zeros((N, 3))
    
    for i in prange(N):  # Use parallel range
        for j in range(N):
            if i != j:
                r_ij = X[j] - X[i]
                dist = np.linalg.norm(r_ij)
                if dist > 0:
                    a[i] += r_ij / dist**3
                    
    return a
```

```python
def measure_parallel_scaling(N_values):
    times_numba = []
    times_parallel = []

    for N in N_values:
        X = np.random.randn(N, 3)

        # Parallel Numba timing
        start_time = time.perf_counter()
        get_acceleration_parallel(X)  # Use compiled function here
        end_time = time.perf_counter()
        times_parallel.append(end_time - start_time)

    # Clip very small times_parallel values to avoid divide by zero issues
    times_parallel = np.clip(times_parallel, 1e-10, None)

    return times_parallel
```

```python
N_values = [10, 20, 50, 100, 200, 500, 1000]
times_parallel = measure_parallel_scaling(N_values)
times_numba = measure_execution_time(N_values, get_acceleration_numba)
```

```python
# Plot the performance results
def plot_parallel_scaling(N_values, times_numba, times_parallel):
    plt.figure()
    plt.loglog(N_values, times_numba, marker='o', linestyle='-', label='Numba (non-parallel)')
    plt.loglog(N_values, times_parallel, marker='s', linestyle='-', label='Numba (parallel)')
    plt.xlabel('Number of bodies (N)')
    plt.ylabel('Execution Time (s)')
    plt.legend()
    plt.title('Parallel vs. Non-Parallel Performance')
    plt.show()
```

```python
N_values = [10, 20, 50, 100, 200, 500, 1000]

for N in N_values:
    X = np.random.rand(N, 3)  # Random positions for N bodies
    print(f"\nValidating for N={N}:")
    validate_acceleration(get_acceleration_numba, get_acceleration_parallel, X)
```

```python
plot_parallel_scaling(N_values, times_numba, times_parallel)
speedup_factors = np.array(times_numba) / np.array(times_parallel)
print("Speedup Factors (Numba Non-Parallel / Numba Parallel):", speedup_factors)
```

We used the same function for computing accelerations as before, but this time we modified our "get_acceleration" function to run in Numba's "prange". This allows different CPU core to compute accelerations for multiple bodies simultaneously, instead of processing them one at a time.

To do that we define "get_acceleration_parallel" and add "parallel=True" to @njit. We replace the outer loop with "prange" instead of "range". This enables parallel execution, where different iterations of the loop run on separate CPU threads. The inner loop remains unchanged since each body still needs to interact with all others. The result is that each bodyâs acceleration is computed independently, allowing multiple computations at once. To ensure the parallelized function still gives correct results, we use our validate_acceleration function. We compare their outputs using our manual validation function and allclose. The validation confirms that Numba parallelization does not affect correctnessâonly execution speed.

To measure execution time, we used the same timing function as before but separately measured Numba (non-parallel) and Numba (parallelised) inside the loop. This allowed us to directly compare their performance.

Our performance plot shows that the parallelized version is consistently faster than the standard Numba implementation, achieving speedup factors ranging from  $2$ to $9$ times. The improvement depends on N and the number of available CPU cores. However for $N=10$ bodies takes far longer for the parallelised function, than for the non-parallelised which seems odd. Similiarly for $N=20$ bodies, the execution time seems to be the same. This happens because of parallelization overhead. When running in parallel, the system must split the work among multiple CPU threads. For small $N$, the computation itself is very fast, but creating and managing threads takes extra time. This overhead can outweigh the benefits of parallelization when the workload is small.

```python
def weak_scaling_test(N_values):
    times = []
    for N in N_values:
        X = np.random.randn(N, 3)
        start_time = time.perf_counter()
        get_acceleration_parallel(X)
        end_time = time.perf_counter()
        times.append(end_time - start_time)
    
    plt.figure()
    plt.plot(N_values, times, marker='o', linestyle='-', label='Parallel Weak Scaling')
    plt.xlabel('Number of Bodies (N)')
    plt.ylabel('Execution Time (s)')
    plt.legend()
    plt.title('Weak Scaling Test')
    plt.show()
```

```python
weak_scaling_test(N_values)
```

The weak scaling test evaluates how well the parallel implementation performs as the problem size $N$ increases while keeping the computational work per core constant. Ideally, if parallelization is efficient, execution time should remain nearly constant as $N$ increases, since the additional workload is evenly distributed across multiple CPU cores.

To test this, we ran our parallelized function for increasing values of $N$ and recorded the execution times. A perfectly efficient parallel implementation would result in a flat curve, meaning execution time remains unchanged despite increasing problem size.

Our results show that execution time only slightly increases, with the largest jump occurring between $N=500$ and $N=1000$, where the time increases by approximately $0.03$ seconds. This indicates that the parallel implementation scales reasonably well, but some performance overheadâsuch as memory access latency, thread synchronization, or CPU cache limitationsâmay start affecting efficiency at larger $N$.

```python
def strong_scaling_test(N_fixed, num_threads_list):
    times = []
    X = np.random.randn(N_fixed, 3)

    for num_threads in num_threads_list:
        numba.set_num_threads(num_threads)
        start_time = time.perf_counter()
        get_acceleration_parallel(X)
        end_time = time.perf_counter()
        times.append(end_time - start_time)
    
    plt.figure()
    plt.plot(num_threads_list, times, marker='o', linestyle='-', label='Parallel Strong Scaling')
    plt.xlabel('Number of Threads')
    plt.ylabel('Execution Time (s)')
    plt.legend()
    plt.title('Strong Scaling Test')
    plt.show()
```

```python
# Define number of threads to test
thread_values = [1,2,3,4,5,6,7,8]  # Adjust based on CPU

# Set a fixed problem size for strong scaling
N_fixed = 1000

# Run Strong Scaling Test
times_strong = strong_scaling_test(N_fixed, thread_values)
```

The strong scaling test measures how efficiently the parallelized function benefits from increasing the number of CPU threads while keeping the problem size (N_fixed) constant. Ideally, if parallelization is perfectly efficient, doubling the number of threads should cut execution time in half, indicating a linear speedup. 

To test this, we fixed the problem size (N_fixed$=1000$) so that the total computational workload remains constant. We then varied the number of CPU threads from $1$ to $8$ and measured execution times for each case.

If parallelization is effective, the execution time should decrease as more threads are used. Our results confirm this: as we increase the number of threads, execution time steadily decreases, demonstrating that the parallel implementation successfully distributes work across multiple cores. However, at some point, scaling might slow down due to factors like thread overhead, memory bandwidth limitations, or diminishing workload per thread.
