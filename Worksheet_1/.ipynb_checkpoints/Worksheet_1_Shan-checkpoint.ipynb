```python
import numpy as np
import time
import matplotlib.pyplot as plt
import scipy
import matplotlib
import sympy
import numba
import jax
import mpi4py
from numba import njit, prange , set_num_threads, get_num_threads
from typing import Callable
```

```python
def get_acceleration(X: np.ndarray) -> np.ndarray:
    """
    Compute gravitational acceleration for a system of N bodies.
    
    Parameters:
        X (np.ndarray): 2D array of shape (N, 3) containing the positions of the bodies.
    
    Returns:
        np.ndarray: 2D array of shape (N, 3) containing the accelerations of the bodies.
    """
    N = X.shape[0]
    A = np.zeros_like(X)  # Initialize acceleration array
    
    for i in range(N):
        for j in range(N):
            if i != j:
                r_ij = X[j] - X[i]  # Vector from i to j
                distance = np.linalg.norm(r_ij)
                if distance > 0:
                    A[i] += r_ij / distance**3  # Gravitational acceleration formula
    
    return A
```

```python
def measure_execution_time(N_values):
    times = []
    for N in N_values:
        X = np.random.randn(N, 3)  # Generate random positions
        start_time = time.time()
        get_acceleration(X)
        end_time = time.time()
        times.append(end_time - start_time)
    return times

def plot_scaling(N_values, times):
    plt.figure()
    plt.loglog(N_values, times, marker='o', linestyle='-', label='Measured time')
    plt.loglog(N_values, (np.array(N_values)**2) * (times[0] / N_values[0]**2), linestyle='--', label='O(N²) reference')
    plt.xlabel('Number of bodies (N)')
    plt.ylabel('Execution Time (s)')
    plt.legend()
    plt.title('Scaling of Gravitational Acceleration Computation')
    plt.show()

# Run and plot results
N_values = [10, 20, 50, 100, 200, 500]  # Different values of N to test
execution_times = measure_execution_time(N_values)
plot_scaling(N_values, execution_times)
```

####  3. Write the same function using (non-parallel) Numba and compare the performance of the numpy and numba-based functions. What is the speedup factor (if any)? Repeat the timing measurements for different values of N and insert the results into the plot from the previous question 
### Steps:
Use @njit from Numba to optimize the function \
Compare the performance of NumPy vs. Numba implementations. \
Measure speedup for different values of 𝑁. \
Plot both execution times on the same graph.

```python
# Optimized Numba-based function
@njit
def get_acceleration_numba(X: np.ndarray) -> np.ndarray:
    """
    Compute gravitational acceleration for a system of N bodies using Numba.
    """
    N = X.shape[0]
    A = np.zeros_like(X)  # Initialize acceleration array

    for i in range(N):
        for j in range(N):
            if i != j:
                r_ij = X[j] - X[i]  # Vector from i to j
                distance = np.linalg.norm(r_ij)
                if distance > 0:
                    A[i] += r_ij / distance**3  # Gravitational acceleration formula
    return A
get_acceleration_numba(np.random.randn(10, 3))  # Warm-up call

# Function to measure execution time for both methods
def measure_execution_time_comparison(N_values):
    times_numpy = []
    times_numba = []

    for N in N_values:
        X = np.random.randn(N, 3)  # Generate random positions

        # Measure NumPy execution time
        start_time = time.time()
        get_acceleration(X)
        end_time = time.time()
        times_numpy.append(end_time - start_time)

        # Measure Numba execution time
        start_time = time.time()
        get_acceleration_numba(X)  # First call may take longer due to JIT compilation
        end_time = time.time()
        times_numba.append(end_time - start_time)

    return times_numpy, times_numba
```

```python
# Plot performance comparison
def plot_performance_comparison(N_values, times_numpy, times_numba):
    plt.figure()
    plt.loglog(N_values, times_numpy, marker='o', linestyle='-', label='NumPy')
    plt.loglog(N_values, times_numba, marker='s', linestyle='-', label='Numba')
    plt.xlabel('Number of bodies (N)')
    plt.ylabel('Execution Time (s)')
    plt.legend()
    plt.title('NumPy vs. Numba Performance')
    plt.show()

# Run and compare results
N_values = [10, 20, 50, 100, 200, 500]

times_numpy, times_numba = measure_execution_time_comparison(N_values)
plot_performance_comparison(N_values, times_numpy, times_numba)

# Compute Speedup Factor
speedup_factors = np.array(times_numpy) / np.array(times_numba)
print("Speedup Factors (NumPy / Numba):", speedup_factors)
```

NumPy (blue) takes significantly longer than Numba (orange) for large 𝑁 , confirming that the naive NumPy implementation is much slower. \
Numba (orange) is much faster for large 𝑁, which is expected due to JIT compilation. \

#### 4  Write a function that validates whether the two functions returned the same results.

```python
def validate_acceleration(method1: Callable, method2: Callable, X: np.ndarray) -> bool:
    """
    Compare the results of two acceleration computation methods.

    Parameters:
        method1 (Callable): First function to compute acceleration.
        method2 (Callable): Second function to compute acceleration.
        X (np.ndarray): Input array of positions of shape (N, 3).

    Returns:
        bool: True if the results are identical within numerical tolerance, False otherwise.
    """
    A1 = method1(X)
    A2 = method2(X)
    
    return np.allclose(A1, A2, atol=1e-6)  # Allow for minor floating-point differences

validation_results = {N: validate_acceleration(get_acceleration, get_acceleration_numba, np.random.randn(N, 3)) for N in N_values}

# Print the dictionary of results
print(validation_results)
```

### 5. Write a parallel version of the get acceleration function using Numba and verify correctness of the results. Perform a weak and a strong scaling test.

```python
# Parallelized Numba-based function
@njit(parallel=True)
def get_acceleration_parallel(X: np.ndarray) -> np.ndarray:
    """
    Compute gravitational acceleration for a system of N bodies using parallel Numba.
    
    Parameters:
        X (np.ndarray): 2D array of shape (N, 3) containing the positions of the bodies.
    
    Returns:
        np.ndarray: 2D array of shape (N, 3) containing the accelerations of the bodies.
    """
    N = X.shape[0]
    A = np.zeros_like(X)  # Initialize acceleration array

    for i in prange(N):  # Use prange for parallel execution
        for j in range(N):
            if i != j:
                r_ij = X[j] - X[i]  # Vector from i to j
                distance = np.linalg.norm(r_ij)
                if distance > 0:
                    A[i] += r_ij / distance**3  # Gravitational acceleration formula
    return A
get_acceleration_parallel(np.random.randn(10, 3))  # Warm-up call


# Perform timing tests for weak and strong scaling
def measure_parallel_scaling(N_values):
    times_numba = []
    times_parallel = []

    for N in N_values:
        X = np.random.randn(N, 3)

        # Numba (non-parallel) timing
        start_time = time.time()
        get_acceleration_numba(X)
        end_time = time.time()
        times_numba.append(end_time - start_time)

        # Parallel Numba timing
        start_time = time.time()
        get_acceleration_parallel(X)  # Ensure JIT compilation before timing
        end_time = time.time()
        times_parallel.append(end_time - start_time)

    return times_numba, times_parallel
```

```python

# Run the scaling tests
times_numba, times_parallel = measure_parallel_scaling(N_values)

# Plot the performance results
def plot_parallel_scaling(N_values, times_numba, times_parallel):
    plt.figure()
    plt.loglog(N_values, times_numba, marker='o', linestyle='-', label='Numba (non-parallel)')
    plt.loglog(N_values, times_parallel, marker='s', linestyle='-', label='Numba (parallel)')
    plt.xlabel('Number of bodies (N)')
    plt.ylabel('Execution Time (s)')
    plt.legend()
    plt.title('Parallel vs. Non-Parallel Performance')
    plt.show()

plot_parallel_scaling(N_values, times_numba, times_parallel)

# Compute Speedup Factor
speedup_factors = np.array(times_numba) / np.array(times_parallel)
print("Speedup Factors (Numba Non-Parallel / Numba Parallel):", speedup_factors)
```
